```python
import cv2

# Load Haar Cascade
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
)

video_cap = cv2.VideoCapture(1)  # use correct camera index

while True:
    ret, frame = video_cap.read()
    
    if not ret:
        break

    # Convert to grayscale (required)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = face_cascade.detectMultiScale(
        gray,
        scaleFactor=1.3,
        minNeighbors=5
    )

    # Draw rectangle + text
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

        cv2.putText(
            frame,
            "Face Detected",
            (x, y - 10),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.7,
            (0, 255, 0),
            2
        )

    cv2.imshow("Face Detection", frame)

    if cv2.waitKey(1) & 0xFF == 27:  # ESC key
        break

video_cap.release()
cv2.destroyAllWindows()

for _ in range(5):
    cv2.waitKey(1)
```

```python
import cv2
import sys

# Load Haar Cascade
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
)

video_cap = cv2.VideoCapture(1)  # change index if needed

while video_cap.isOpened():
    ret, frame = video_cap.read()

    if not ret:
        print("Failed to grab frame")
        break

    # Resize (optional for performance)
    frame = cv2.resize(frame, (640, 480))

    # Convert to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = face_cascade.detectMultiScale(
        gray,
        scaleFactor=1.3,
        minNeighbors=5
    )

    # Draw rectangle + text
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

        cv2.putText(
            frame,
            "Face Detected",
            (x, y - 10),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.7,
            (0, 255, 0),
            2
        )

    cv2.imshow("Face Detection", frame)

    # 🔑 KEY FIX: capture key properly
    key = cv2.waitKey(1) & 0xFF

    # Press ANY key to exit
    if key != 255:   # 255 means "no key pressed"
        break

# Release camera
video_cap.release()

# Destroy windows
cv2.destroyAllWindows()

# 🔥 macOS fix (VERY IMPORTANT)
for _ in range(5):
    cv2.waitKey(1)

# Force close process (removes loading cursor)
sys.exit()
```
